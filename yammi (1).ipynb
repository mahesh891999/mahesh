```python
df['Date'] = pd.to_datetime(df['Date'])

# Plotting
x = df['Date']
y = df['Close']
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title("Stock Market")
plt.xlabel("Date")
plt.ylabel("Price")

# Function to display date on hover
def on_plot_hover(event):
    for line in plt.gca().get_lines():
        if line.contains(event)[0]:
            x_val = event.xdata
            y_val = event.ydata
            index = list(x).index(x_val)
            date_str = df.iloc[index]['Date'].strftime('%Y-%m-%d')
            plt.gca().annotate(f'{date_str}', xy=(x_val, y_val), xytext=(x_val, y_val + 5),
                               ha='center', textcoords='offset points', fontsize=8,
                               bbox=dict(boxstyle='round,pad=0.2', fc='yellow', alpha=0.8))
            plt.draw()

plt.gcf().canvas.mpl_connect('motion_notify_event', on_plot_hover)

plt.tight_layout()
plt.show()
```
